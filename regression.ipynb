```python
data = pd.concat([data_jan, data_feb, data_mar, data_apr, data_may, data_jun, data_jul, data_aug, data_sep, data_oct, data_nov, data_dec], ignore_index=True)
```

```python
data['asset_description'] = data['asset_description'].astype(str)
data['asset_collection_description'] = data['asset_collection_description'].astype(str)
```

```python
new_column = []

for index, row in data.iterrows():
    if row['asset_description'] == '1' or row['asset_collection_description'] == '1':
        new_column.append(1)
    else:
        new_column.append(0)

data['generator'] = new_column
```

```python
sum(data['generator'])
```

```python
sum(data['generator']) / len(data)
```

```python
data = data.drop(['asset_description','asset_collection_description', 'event_timestamp', 'asset_collection_created_date', 'transaction_timestamp'], axis=1)
```

```python
numbers = [sum(data['generator']), len(data) - sum(data['generator'])]
labels = ['generative', 'hand-drawn']

# Create a pie chart
plt.pie(numbers, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.show()
```

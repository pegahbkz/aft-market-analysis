```python
data = pd.concat([data_jan, data_feb, data_mar, data_apr, data_may, data_jun, data_jul, data_aug, data_sep, data_oct, data_nov, data_dec], ignore_index=True)
```

```python
data['asset_description'] = data['asset_description'].astype(str)
data['asset_collection_description'] = data['asset_collection_description'].astype(str)
```

```python
new_column = []

for index, row in data.iterrows():
    if row['asset_description'] == '1' or row['asset_collection_description'] == '1':
        new_column.append(1)
    else:
        new_column.append(0)

data['generator'] = new_column
```

```python
data = data.drop(['asset_description','asset_collection_description', 'event_timestamp', 'asset_collection_created_date', 'transaction_timestamp'], axis=1)
```

```python
df2 = pd.DataFrame(columns=data.columns)

# Store the dtypes of original_df columns
dtypes_dict = data.dtypes.to_dict()

# Iterate through the original DataFrame
indices_to_remove = []  # To store indices of rows to be removed
for index, row in data.iterrows():
    if row['generator'] == 1:
        # Add rows where 'a' is equal to 1 to df2
        df2 = df2.append(row, ignore_index=True)
        indices_to_remove.append(index)  # Store indices to remove

# Explicitly cast the types in df2 to match original_df
for col, dtype in dtypes_dict.items():
    df2[col] = df2[col].astype(dtype)
    
data = data.drop(indices_to_remove)
```

```python
df3 = data.sample(frac=0.12, random_state=42)
```

```python
df = pd.concat([df2, df3], ignore_index=True)
```

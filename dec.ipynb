```python
from opensea import utils as opensea_utils
import json
import csv
from flatten_json import flatten
```

```python
count = 0

for j in range(24):
    period_start = opensea_utils.datetime_utc(2023, 12, 1, j, 0)
    period_end = opensea_utils.datetime_utc(2023, 12, 1, j, 59)

    result = api.events(
        occurred_after=period_start,
        occurred_before=period_end,
        limit=300,
        export_file_name="23-dec1.json",
        event_type="successful",
)

    with open('23-dec1.json') as json_file:
        jsondata = json.load(json_file)


        data_file = open('23-dec.csv', 'a', newline='')
        csv_writer = csv.writer(data_file)

        count = 0

        for data in jsondata["asset_events"]:
            data = flatten(data)
            if count == 0:
                header = data.keys()
                csv_writer.writerow(header)
                count += 1
            csv_writer.writerow(data.values())

        data_file.close()
```
